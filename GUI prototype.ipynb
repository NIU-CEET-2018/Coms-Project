```python
%gui qt5
import sys
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *





class Example(QMainWindow):
    
    def __init__(self):
        super().__init__()
        
        self.initUI()
        
        
    def initUI(self):  
        #Status Bar - Bottom left of window - he has to go first
        self.statusBar().showMessage('Running')
        
         #Sets the central widget to be a text editor
        self.textEdit = QTextEdit()
        self.setCentralWidget(self.textEdit)
        
        #Exit Act from QAction
        exitAct = QAction(QIcon('exit.png'), '&Exit', self)
        exitAct.setShortcut('Ctrl+Q')
        exitAct.setStatusTip('Exit application')
        exitAct.triggered.connect(qApp.quit)
        
        #Menu
        menubar = self.menuBar()
        
        #File Menu - Read and Write files
        fileMenu = menubar.addMenu('File')
        
        #New
        newFile = QAction('New', self)
        
        #Open a saved file:
        openFile = QAction(QIcon('open.png'), 'Open', self)
        #openFile.setShortcut('TBA')
        openFile.triggered.connect(self.showDialog)
        
        saveFile = QAction('Save', self)
        saveAs = QAction('Save As', self)
        
        fileMenu.addAction(newFile)
        fileMenu.addAction(openFile)
        fileMenu.addAction(saveFile)
        fileMenu.addAction(saveAs)
        fileMenu.addAction(exitAct)
        
        #Training Menu
        trainingMenu = menubar.addMenu('Train')
        trainingSession = QAction('Run Training Session', self)
        newGesture = QAction('Add a new gesture', self)
        trainingMenu.addAction(trainingSession)
        trainingMenu.addAction(newGesture)
        
        
        #Autocomplete 1, Print out string from autocomplete program
        autocompleteOne = menubar.addMenu('Autcomplete1')
        autocompleteOne.setStatusTip('Prints guessed output from autocomplete module')
        
        #Autocomplete 2
        autocompleteTwo = menubar.addMenu('Autcomplete2')
        autocompleteTwo.setStatusTip('Prints guessed output from autocomplete module')
        
        #Autocomplete 1
        autocompleteThree = menubar.addMenu('Autcomplete1')
        autocompleteThree.setStatusTip('Prints guessed output from autocomplete module')
        
        #Tools Menu
        toolsMenu = menubar.addMenu('Tools')
        showStats = QAction('Statistics', self)
        leapVisualizer = QAction('Leap Visualizer', self)
        toolsMenu.addAction(showStats)
        toolsMenu.addAction(leapVisualizer)
        
        #Storage Menu
        storageMenu = menubar.addMenu('Storage')
        storageToggle = QAction('Storage', self, checkable=True)
        storageToggle.setShortcut('Ext. Switch')
        storageToggle.setStatusTip('Storage ON')
        storageToggle.triggered.connect(self.toggleMenu)
        
        storageMenu.addAction(storageToggle)
        
        #Voice Options
        voiceMenu = menubar.addMenu('Voice')
        
        #Submenu to switch between voices, will also play a sample on select
        voiceChanger = QMenu('Voice Changer', self)
        voiceCortana = QAction('Cortana', self)
        voiceMSSam = QAction('Microsoft Sam', self)
        voiceSiri = QAction('Siri', self)
        voiceVader = QAction('Darth Vader', self)
        voiceAlexa = QAction('Alexa', self)
        
        voiceChanger.addAction(voiceCortana)
        voiceChanger.addAction(voiceMSSam)
        voiceChanger.addAction(voiceSiri)
        voiceChanger.addAction(voiceVader)
        voiceChanger.addAction(voiceAlexa)
        
        voiceMenu.addMenu(voiceChanger)
        
        #Submenu to customize speech prompt
        speechSettings = QMenu('Speech Settings', self)
        speechEnd = QAction('Speech on End of Sentence', self, checkable=True)
        speechCustom = QAction('Customize Speech Prompt', self, checkable=True)
        speechToggle = QAction('Turn Speech Off', self, checkable=True)
        
        speechSettings.addAction(speechEnd)
        speechSettings.addAction(speechCustom)
        speechSettings.addAction(speechToggle)
        
        voiceMenu.addMenu(speechSettings)
        
#Geometry, Naming and positioning       
        self.resize(1024, 600)        
        self.setWindowTitle('Menu')
        self.center()
        self.show()
  
    def toggleMenu(self, state):
        if state:
            self.statusbar().show()
        else:
            self.statusbar().hide()
                          
                          
   #Allows user to navigate through the file directory and open a file                       
    def showDialog(self):
        
        fname = QFileDialog.getOpenFileName(self, 'Open file', '/home')
        
        if fname[0]:
            f = open(fname[0], 'r')
            
            with f:
                data = f.read()
                self.textEdit.setText(data)
        
#closeEvent
    def closeEvent(self, event):
        
        reply = QMessageBox.question(self, 'Message',
            "Are you sure to quit?", QMessageBox.Yes | 
            QMessageBox.No, QMessageBox.No)

        if reply == QMessageBox.Yes:
            event.accept()
        else:
            event.ignore()     

 #Centers the window on the screen           
    def center(self):
        
        qr = self.frameGeometry()
        
        #Finds the center of the screen from QDesktopWidget
        cp = QDesktopWidget().availableGeometry().center()
        
        #Centers on the center of the screen
        qr.moveCenter(cp)
        self.move(qr.topLeft())
        
if __name__ == '__main__':
    
    app = QApplication(sys.argv)
    ex = Example()
    sys.exit(app.exec_())
```
